```python
pd.read_html('https://338canada.com/polls.htm')[0]
df=pd.read_html('https://338canada.com/polls.htm')[0]
df
df["Date (middle)"] = pd.to_datetime(df["Date (middle)"])
df2=df.iloc[0:15]
df2['PPC'] = df2['PPC'].fillna(0)
df2
```

```python
seats = 338
N = df2['N'].sum()
percentLPC = 0
percentCPC = 0
percentNDP = 0
percentGPC = 0
percentBQ = 0
percentPPC = 0
today = date.today()
diff = 0
total = 0
N
```

```python
for i in range(len(df2)) :
    #print(df.loc[i, "LPC"], df.loc[i, "CPC"])
    Poll = df2.loc[i,"LPC"]
    n = df2.loc[i,"N"]
    weighted_total = Poll*(n/N)
    percentLPC = percentLPC + weighted_total/100


for i in range(len(df2)) :
    Poll = df2.loc[i,"CPC"]
    n = df2.loc[i,"N"]
    weighted_total = Poll*(n/N)
    percentCPC = percentCPC + weighted_total/100

for i in range(len(df2)) :
    Poll = df2.loc[i,"NDP"]
    n = df2.loc[i,"N"]
    weighted_total = Poll*(n/N)
    percentNDP = percentNDP + weighted_total/100


for i in range(len(df2)) :
    Poll = df2.loc[i,"GPC"]
    n = df2.loc[i,"N"]
    weighted_total = Poll*(n/N)
    percentGPC = percentGPC + weighted_total/100


for i in range(len(df2)) :
    Poll = df2.loc[i,"BQ"]
    n = df2.loc[i,"N"]
    weighted_total = Poll*(n/N)
    percentBQ = percentBQ + weighted_total/100

    
for i in range(len(df2)) :
    Poll = df2.loc[i,"PPC"]
    n = df2.loc[i,"N"]
    weighted_total = Poll*(n/N)
    percentPPC = percentPPC + weighted_total/100


total = percentLPC + percentCPC + percentNDP+percentGPC+percentBQ+percentPPC
weightedLPC = percentLPC/total
weightedCPC = percentCPC/total
weightedNDP = percentNDP/total
weightedGPC = percentGPC/total
weightedBQ = percentBQ/total
weightedPPC = percentPPC/total

seatsLPC = seats*weightedLPC
seatsCPC = seats*weightedCPC
seatsNDP = seats*weightedNDP
seatsGPC = seats*weightedGPC
seatsBQ = seats*weightedBQ
seatsPPC = seats*weightedPPC
    
print(seatsLPC, seatsCPC, seatsNDP, seatsGPC, seatsBQ, seatsPPC)

#for i in range(len(df2)) :
    #poll_date = df2.loc[i,"Date (middle)"]
    #diff - poll_date - today
```
