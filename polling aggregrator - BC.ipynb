```python
total_seats = 42
tracker_LPC = 0
tracker_CPC = 0
tracker_NDP = 0
tracker_GPC = 0
tracker_BQ = 0
tracker_PPC = 0
```

```python
for i in range(len(df3)) :
    Poll = df3.loc[i,"LPC"]
    tracker_LPC = Poll + tracker_LPC
    percentLPC = tracker_LPC
    
for i in range(len(df3)) :
    Poll = df3.loc[i,"CPC"]
    tracker_CPC = Poll + tracker_CPC
    percentCPC = tracker_CPC
    
for i in range(len(df3)) :
    Poll = df3.loc[i,"NDP"]
    tracker_NDP = Poll + tracker_NDP
    percentNDP = tracker_NDP
    
for i in range(len(df3)) :
    Poll = df3.loc[i,"GPC"]
    tracker_GPC = Poll + tracker_GPC
    percentGPC = tracker_GPC
    
for i in range(len(df3)) :
    Poll = df3.loc[i,"BQ"]
    tracker_BQ = Poll + tracker_BQ
    percentBQ = tracker_BQ
    
for i in range(len(df3)) :
    Poll = df3.loc[i,"PPC"]
    tracker_PPC = Poll + tracker_PPC
    percentPPC = tracker_PPC
   


total = percentLPC + percentCPC +percentNDP +percentGPC +percentPPC +percentBQ 
weightedLPC = percentLPC/total
weightedCPC = percentCPC/total
weightedNDP = percentNDP/total
weightedGPC = percentGPC/total
weightedBQ = percentBQ/total
weightedPPC = percentPPC/total

seatsLPC = total_seats*weightedLPC
seatsCPC = total_seats*weightedCPC
seatsNDP = total_seats*weightedNDP
seatsGPC = total_seats*weightedGPC
seatsBQ = total_seats*weightedBQ
seatsPPC = total_seats*weightedPPC
    
print(seatsLPC,seatsCPC, seatsNDP, seatsGPC, seatsBQ, seatsPPC)
```
